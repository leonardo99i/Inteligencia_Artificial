# Introdução

Existem dois tipos de aprendizados que são utilizados no *Machine Learning*. São eles o aprendizidado supervisionado e o aprendizado não supervisionado.

*   No aprendizado supervisionado, o modelo é treinando com dados rotulados, ou seja, dados que já possuem uma "resposta" esperada, para que o modelo posua uma base.

*   No aprendizado não supervisionado, o modelo é treinado com dados não rotulados. Este tipo de aprendizado também é conhecido como *clustering*, pois o foco aqui é agrupar os dados em *clusters*. 


O foco da aula de hoje é no **aprendizado supervisionado**.

-----------------------------------------------------------

O aprendizado supervisionado por sua vez pode ser dividido em dois tipos de tarefas. São elas a classificação e a regressão.

*   A classificação trabalha com um conjunto de dados com o propósito de classificar **variáveis discretas**. Pode-se por exemplo analisar diversos dados de transações bancárias e com eles identificar se o cliente em questão tem risco de dar um calote ou não.

*   A regressão trabalha com um conjunto de dados com o propósito de obter uma ou mais saídas que possuam relação com esses dados. A saída em questão geralmente é uma **variável contínua**. Pode-se por exemplo tentar prever a concentração de determinado gás de efeito estufa em uma região.


-----------------------------------------------------------

Procedimento de aprendizado supervisionado:


*   Selecionar o dataset que será utilizado
*   Fazer uma análise exploratória dos dados e realizar pré-processamento dos dados se necessário
*   Escolher um modelo de aprendizado supervisionado
*   Realizar o treinamento do modelo em questão
*   Avaliar o desempenho do modelo (coletar métricas) e análise crítica


-----------------------------------------------------------

# Regressão Linear

image.png

A regressão linear geralmente vai utilizar uma ou mais variáveis discretas (X1, X2, ..., Xn) para prever o comportamento de uma outra variável discreta (Y).

No caso onde utiliza-se uma variável (X) para prever o comportamento da outra (Y) , o modelo irá gerar uma reta, conforme a figura acima.

A reta será gerada conforme a equação:

Y = a*x + b

Ou seja, com base nos dados fornecidos, o modelo irá buscar o valor ideal de *a* e *b*.

Caso mais de uma variável de entrada seja utilizada (ex: X1 e X2), a regressão linear passa a ser chamada de regressão linear múltipla.

Ao invés de uma reta, uma curva será gerada, conforme a seguinte equação:

Y = a1*x1 + a2*x2 + ... + an*xn + b



Algumas Métricas de Desempenho de Modelos de Regressão

*   MAE (*Mean Absolute Error*)

*   RMSE (*Root Mean Square Error*)

## Utilizando Biblioteca Pronta

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Documentação: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_regression.html
# O noise representa um valor de desvio padrão que vai "espalhar" os dados
x, y = make_regression(n_samples = 100, n_features = 1, noise = 10)

plt.scatter(x,y)
plt.show()
```

```python
# Documentação: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

model = LinearRegression()
model.fit(x,y)
```

```python
a = model.coef_
b = model.intercept_

print("a = ", a)
print("b = ", b)
```

```python
reta = a*x + b
```

```python
plt.scatter(x,y)
plt.plot(x, reta, color="red")
plt.show()
```

```python
MAE = mean_absolute_error(y, reta)
RSME = np.sqrt(mean_squared_error(y, reta))

print("MAE: ", MAE)
print("RSME: ", RSME)
```

Aqui iremos utilizar um exemplo "simples" de reconhecimento de dígitos. Esse exemplo também é bastante utilizado como "Hello World" de Deep Learning.

Carregando o dataset da biblioteca

Teste do modelo

```python
previsoes = model.predict(x_test)

previsoes
```

Verificando a acurácia do modelo

```python
from sklearn.metrics import accuracy_score

acuracia = accuracy_score(y_test,previsoes)
print(acuracia)
```

Pré-processamento dos dados

```python
# Dropando a primeira coluna, pois ela nao tem informacoes relevantes
df.drop(['Sample Code Number'], axis=1, inplace=True) 

# Dropando a colun "Bare Nuclei"
df.drop(['Bare Nuclei'], axis=1, inplace=True)

# Mudando a representacao de "Classe" de 2/4 para 0/1
# Agora 0 indica um cancer benigno e 1 indica um cancer maligno
df['Class'] = np.where(df['Class'] == 2,0,1)
```

```python
df['Class'].value_counts()
```

Separando o dataset em teste/treino

```python
# X são todas as columas, com exceção da coluna de Classe
X = df.drop(['Class'],axis=1)

# Y é apenas a coluna de Classe
y = df['Class']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

Treinamento do modelo

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

Teste do modelo

```python
y_pred = model.predict(X_test)

y_pred
```

Mais métricas que podem ser verificadas

image.png

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
print('Acuracia : ' + str(accuracy_score(y_test,y_pred)))
print('Precisao : ' + str(precision_score(y_test,y_pred)))
print('Recall : ' + str(recall_score(y_test,y_pred)))
print('F1 Score : ' + str(f1_score(y_test,y_pred)))
```

Matriz de confusão

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(str(cm))
```

Apresentando as informações da matriz de confusão de uma melhor forma

```python
plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Classe Real');
plt.xlabel('Classe Prevista');
plt.show()
```
